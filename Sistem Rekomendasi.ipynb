```
# Impor library yang dibutuhkan
import pandas as pd
import numpy as np
import ast
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```
# Memasukkan data csv untuk diolah
df = pd.read_csv('Kerajinan Rajut Final.csv')
```

```
# Contoh 5 data teratas dari tabel
df.head()
```

```
# Merubah data pada tabel warna, kesulitan, kategori, jenis_benang, ukuran_benang menjadi list
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['nama'])
    return L
```

```
# Menerapkan fungsi convert pada kolom warna, kesulitan, kategori, jenis_benang, ukuran_benang
df['warna'] = df['warna'].apply(convert)
df['kesulitan'] = df['kesulitan'].apply(convert)
df['kategori'] = df['kategori'].apply(convert)
df['jenis_benang'] = df['jenis_benang'].apply(convert)
df['ukuran_benang'] = df['ukuran_benang'].apply(convert)
```

```
# Hasil Tabel setelah dilakukan perubahan
df.head()
```

```
# Menghilangkan kolom id
crochets = df.drop(columns=['id'])
```

```
crochets.head()
```

```
# Menggabungkan kolom warna, kesulitan, kategori, jenis_benang, ukuran_benang menjadi satu kolom (tags)
crochets['tags'] = crochets['warna'] + crochets['kesulitan'] + crochets['kategori'] + crochets['jenis_benang'] + crochets['ukuran_benang']

crochets.head()
```

```
# Menghilangkan kolom warna, kesulitan, kategori, jenis_benang, ukuran_benang
crochets = crochets.drop(columns=['warna', 'kesulitan', 'kategori', 'jenis_benang', 'ukuran_benang'])
```

```
# Hasil tabel setelah dilakukan perubahan
crochets.head()
```

```
# Mengubah kolom tags menjadi string dengan menghapus semua koma (,)
crochets['tags'] = crochets['tags'].apply(lambda x: " ".join(x))
```

```
# Hasil tabel setelah dilakukan perubahan
crochets.head()
```

```
# Menggunakan CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```
# Mengubah kolom tags menjadi array
vectors = cv.fit_transform(crochets['tags']).toarray()
```

```
# Hasil kolom tags dalam bentuk array
vectors
```

```
# Menghitung cosine similarity untuk mendapatkan rekomendasi
similarity = cosine_similarity(vectors)
```

```
# Hasil cosine similarity
similarity
```

```
# Membuat fungsi untuk mendapatkan rekomendasi
def recommend(kerajinan):
    index = crochets[crochets['kerajinan'] == kerajinan].index[0]
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])
    for i in distances[1:6]:
        print(crochets.iloc[i[0]].kerajinan)
        print(crochets.iloc[i[0]].link)
```

```
# Hasil Rekomendasi
recommend("Tatakan Calabria")
```
